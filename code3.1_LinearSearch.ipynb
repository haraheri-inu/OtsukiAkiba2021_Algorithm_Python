```python
def linear_search(N: int, v: int):
    """
    input:
    N:int, 探索する配列の長さ
    v:int, 存在するか調べる値
    output:
    'Yes.', 存在する場合
    'No.', 存在しない場合
    """
    # 探索対象のリストを作成
    object_list = []
    for i in range(N):
        num = int(input())
        object_list.append(num)
    # 探索
    exist = False
    for i in range(N):
        if object_list[i] == v:
            exist = True
            break
    if exist == True:
        print('Yes.')
    else: print('No.')
```

```python
# test
linear_search(3, 2)
```

```python
linear_search(10, 4)
```
